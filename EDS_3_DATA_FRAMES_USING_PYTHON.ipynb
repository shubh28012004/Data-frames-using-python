**MULTI INDEX AND INDEX HIERARCHY**

```python
#index levels
outside = ['G1' , 'G1' , 'G1' , 'G2' , 'G2' , 'G2' ]
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside)) #zip is going to collect entries. It  iterate through outside and covert it into tuples
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df #mutiple indexing
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df.index.names
```

```python
df.index.names = ['Group','Num']
```
